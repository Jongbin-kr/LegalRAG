```python
import pandas as pd

# CSV 파일 경로
path='completed/'
filename = '[05월21일03시21분]교통사고고단(2010~2013)'
extension='.csv'
file_path=path+filename+extension
# CSV 파일을 DataFrame으로 불러오기
df = pd.read_csv(file_path)

# 결측치가 있는 특정 열의 이름
columns_to_check = ['범죄사실', '법령의 적용']

# 결측치가 있는 행 제거
df_cleaned = df.dropna(subset=columns_to_check)

# 필요한 열만 남기기
df_cleaned = df_cleaned[columns_to_check]
df_cleaned
```

```python
# 결과 DataFrame을 다시 CSV 파일로 저장 (덮어쓰기)
# df_cleaned.to_csv(file_path, index=False)

# 결과를 새로운 파일로 저장 (원본 보존)
df_cleaned.to_csv(f'{filename}_preprocessed{extension}', index=False)
```
