법령정보와 retrieve된 test data 합치기

```python
import pandas as pd
laws = pd.read_csv('./database/datasets/laws_shuffled.csv')
laws.iloc[655:]
```

```python
df = pd.read_csv('./database/datasets/test_data_with_retrieve_normalized.csv')
df = df.rename(columns={
    'query': '자연어 쿼리',
    'fact': '문서 범죄 사실',
    'laws': '문서 법령의 적용',
    'score': '문서 유사도 점수',
})
df = df.set_index('query_index').reset_index()
df
```

```python
merged_df = pd.merge(df, laws, left_on='document_index', right_on='Unnamed: 0', suffixes=('_df', '_laws'))
merged_df = merged_df.loc[:, ['query_index', '자연어 쿼리', 'document_index', '문서 범죄 사실', '문서 법령의 적용',  '법령명', '법설명', '문서 유사도 점수']]
merged_df = merged_df.rename(columns={
    '법령명': "문서 법령명",
    '법설명': '문서 법설명',
})
merged_df.query_index = merged_df.query_index + 658
merged_df.iloc[::5, :]
```

retrieve된 DB + 법령정보에 쿼리 범죄 사실, 쿼리 법령 정보, 쿼리 법령명 덧붙이기

```python
query = pd.read_csv('./database/datasets/test_data_with_query.csv')
query
```

```python
merged_df2 = pd.merge(merged_df, query, left_on='query_index', right_on='Unnamed: 0')
print(merged_df2.columns, len(merged_df2.columns))

merged_df2 = merged_df2.loc[:, ['query_index', 'Unnamed: 0', '자연어 쿼리_x', '자연어 쿼리_y', '범죄 사실', '법령의 적용', 'document_index', '문서 범죄 사실', '문서 법령의 적용', '문서 법령명', '문서 법설명', '문서 유사도 점수', ]]
print(merged_df2.columns, len(merged_df2.columns))

merged_df2 = merged_df2.loc[:, ['query_index', '자연어 쿼리_x', '범죄 사실', '법령의 적용', 'document_index', '문서 범죄 사실', '문서 법령의 적용', '문서 법령명', '문서 법설명', '문서 유사도 점수']]
merged_df2 = merged_df2.rename(columns={
    '자연어 쿼리_x': '자연어 쿼리',
    '범죄 사실': '쿼리 범죄 사실',
    '법령의 적용': '쿼리 법령의 적용',
})
merged_df2[::5]
```
